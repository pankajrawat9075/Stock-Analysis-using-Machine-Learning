# Importing Libraries for Analysis and Machine Learning

```python
import random
from sklearn.linear_model import LinearRegression

from pandas_datareader import data
from pandas_datareader._utils import RemoteDataError
import seaborn as sns
import numpy as np
import pandas as pd 
from datetime import datetime
```

# Selecting the Particular Stock

```python
stock_list=[]
START_DATE =datetime.date(pd.to_datetime('2000-01-01'))
END_DATE =datetime.date(datetime.now())
STOCK='TCS.NS'
stock_list.append(STOCK)
stock_list
```

```python
def get_data(ticker):
    try:
        stock_data=data.DataReader(ticker,'yahoo',START_DATE,END_DATE)
        return stock_data
    except RemoteDataError:
        print('no dta found for <t>'.format(t=ticker))
```

# Data Collection using the Yahoo FINANCE API

```python
mydata=get_data(STOCK)
mydata.head()
```

```python
mydata.reset_index(inplace=True)
mydata.reset_index(inplace=True)
```

# Additional additional features for better prediction 

```python
no_of_prev_days=50
no_of_next_days=1
prev_days=np.array(list(range(1,no_of_prev_days)))
next_days=np.array(list(range(1,no_of_next_days)))
```

```python
nan=float('NaN')
for i in range(no_of_prev_days):
    data=mydata.Close.tolist()
    for j in range(i+1):
        data.insert(0,nan)
        data.pop()
    mydata[('previous_day '+str(i+1))]=data
for i in range(no_of_next_days):
    data=mydata.Close.tolist()
    for j in range(i+1):
        data.pop(0)
        data.append(nan)
    mydata[('next_day '+str(i+1))]=data
print(mydata.head())
print(mydata.tail())
mydata.columns
```

```python
mydata.drop(['Date','Adj Close'],inplace=True , axis=1)
```

```python
mydata=mydata.dropna()
mydata.columns
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, metrics
```

# Linear Regression Model

# Importing DATA

```python
X = list(mydata.columns)
X.remove('next_day 1')
y = mydata['next_day 1']
X=mydata[X]
```

# Organizing data into training & testing sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)
```

# Model evaluation & prediction

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
print('Coefficients: \n', reg.coef_)
print('Variance score: {}'.format(reg.score(X_test, y_test)))
plt.style.use('fivethirtyeight')
plt.scatter(reg.predict(X_train), reg.predict(X_train) - y_train,
   color = "green", s = 10, label = 'Train data')
plt.scatter(reg.predict(X_test), reg.predict(X_test) - y_test,
   color = "blue", s = 10, label = 'Test data')
plt.hlines(y = 0, xmin = 0, xmax = 50, linewidth = 2)
plt.legend(loc = 'upper right')
plt.title("Residual errors")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
train_preds = reg.predict(X_train)
mse = mean_squared_error(y_train, train_preds)
rmse = sqrt(mse)
rmse
```

```python
test_preds = reg.predict(X_test)
mse = mean_squared_error(y_test, test_preds)
rmse = sqrt(mse)
rmse
```

```python
y_pred = reg.predict(X_test)
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
display(df)
```

```python
X_train.columns
```

```python
X_t = np.array(X_train['index']).reshape(-1, 1)
plt.figure(1, figsize=(16,10))
plt.title('Linear Regression | Price vs Time')
plt.scatter(X_t, y_train, color='blue', label='Actual Price')
plt.scatter(X_t, reg.predict(X_train), color='red', label='Predicted Price')

plt.xlabel('Integer Date')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
```

```python

plt.scatter(y_test, test_preds)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')

plt.title('Predicted Actual Price in future')

plt.show()
```

```python
correct=(abs(y_test-y_pred)<=20).sum()
total=len(y_pred)
(correct/total)*100
```
